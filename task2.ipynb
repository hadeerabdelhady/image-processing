```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Jelly_cc11.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
print(img)
```

```python
hist = cv2.calcHist([gray], [0], None, [256], [0,256])

plt.plot(hist)
plt.title("Original Image Histogram")
plt.xlabel("Gray Level")
plt.ylabel("Number of Pixels")
plt.show()
```

```python
hist = cv2.calcHist([gray], [0], None, [256], [0,256])

r_min = np.where(hist > 0)[0][0]#First non-zero bin return the index of the first non-zero bin
r_max = np.where(hist > 0)[0][-1]#Last non-zero bin return the index of the last non-zero bin

print("r_min =", r_min)
print("r_max =", r_max)
```

```python
stretched = ((gray - r_min) / (r_max - r_min)) * 255
stretched = stretched.astype(np.uint8)
plt.imshow(stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
plt.show()
```

```python
hist2 = cv2.calcHist([stretched], [0], None, [256], [0,256])

plt.plot(hist2)
plt.title("Histogram After Contrast Stretching")
plt.xlabel("Gray Level")
plt.ylabel("Number of Pixels")
plt.show()
```
